```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Loading in the dataset.

file_path = "./Data/monthly_counts_sorted.csv"

monthly_sorted_df = pd.read_csv(file_path)

monthly_sorted_df.head()
```

```python
# Creating a new dataframe that only shows results with the value of "M" in "LAW_CAT_CD" column.

law_cat_m = monthly_sorted_df[monthly_sorted_df['LAW_CAT_CD'] == 'M']

law_cat_m.head()
```

```python
# Creating a new dataframe that only shows results with the value of "F" in "LAW_CAT_CD" column.

law_cat_f = monthly_sorted_df[monthly_sorted_df['LAW_CAT_CD'] == 'F']

law_cat_f.head()
```

```python
# Creating a new dataframe that only shows results with the value of "V" in "LAW_CAT_CD" column.

law_cat_v = monthly_sorted_df[monthly_sorted_df['LAW_CAT_CD'] == 'V']

law_cat_v.head()
```

```python
# Creating a chart to display the data.

fig = px.sunburst(monthly_sorted_df, path=['ARREST_BORO', 'PERP_RACE', 'AGE_GROUP', 'PERP_SEX'], values='counts')

fig.update_layout(width=800, height=800)

fig.show()
```

```python
# Creating a chart to display the data.

fig = px.sunburst(law_cat_m, path=['ARREST_BORO', 'PERP_RACE', 'AGE_GROUP', 'PERP_SEX'], values='counts')

fig.update_layout(width=800, height=800)

fig.show()
```

```python
# Creating a chart to display the data.

fig = px.sunburst(law_cat_f, path=['ARREST_BORO', 'PERP_RACE', 'AGE_GROUP', 'PERP_SEX'], values='counts')

fig.update_layout(width=800, height=800)

fig.show()
```

```python
# Creating a chart to display the data.

fig = px.sunburst(law_cat_v, path=['ARREST_BORO', 'PERP_RACE', 'AGE_GROUP', 'PERP_SEX'], values='counts')

fig.update_layout(width=800, height=800)

fig.show()
```
